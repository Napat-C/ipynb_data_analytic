**6. การอ่านข้อมูลจาก API และ add resource เข้า package**

```python
import requests
from ckanapi import RemoteCKAN
import pandas as pd
import datetime
import numpy as np
from werkzeug.datastructures import FileStorage

r = requests.get('http://api.openweathermap.org/data/2.5/onecall?lat=13.75&lon=100.5167&exclude=current,daily,minutely,alerts&APPID=c69dd4148ac9e1eff9caa0a563a82153')
x = r.json()

df = pd.DataFrame(x['hourly'])
df = df.replace(np.nan, '', regex=True)
df
```

```python
session = requests.Session()
session.verify = False
ua = 'ckanapiexample/1.0 (+http://example.com/my/website)'
portal = RemoteCKAN(ckan_url, user_agent=ua, apikey=api_key, session=session)

tmp_file_name = 'api.csv'
df.to_csv(tmp_file_name, index=False)
f = open(tmp_file_name,"rb")
upload = FileStorage(f,tmp_file_name,name='file',content_type='text/csv')

resource_metadata = {
    #'id': '8296deed-8945-49d7-bb75-340e06215ac9',
    'package_id'    : 'dataset-napat',
    'name': 'Open weather - Bangkok12 Update',
    'format'        : 'CSV',
    'mimetype' : 'text/csv',
    'upload'           : upload,
    'created'       : datetime.datetime.now().isoformat(),
    'last_modified' : datetime.datetime.now().isoformat()
}

if 'id' in resource_metadata:
    print("Update")
    resource = portal.action.resource_update(**resource_metadata)
else:
    print("Create")
    resource = portal.action.resource_create(**resource_metadata)
session.close()
resource
```
