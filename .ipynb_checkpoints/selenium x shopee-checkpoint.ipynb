```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import bs4
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path=r'D:\chorme driver\chromedriver.exe')
```

```python
## get website
driver.get('https://shopee.co.th/')
```

```python
## select language

thai_button = driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[3]/div[1]/button")
```

```python
## click thai button 
thai_button.click()
```

```python
# find close botton location of pop-up in first page
close_button = driver.execute_script('return document.querySelector("shopee-banner-popup-stateful").shadowRoot.querySelector("div.shopee-popup__close-btn")')
```

```python
close_button.click()
```

```python
search = driver.find_element_by_xpath("/html/body/div[1]/div/div[2]/div[1]/div[2]/div/div[1]/div[1]/div/form/input")
```

```python
search.send_keys('โต๊ะปรับระดับไฟฟ้า')
```

```python
search.send_keys(Keys.ENTER)
```

```python
driver.execute_script("document.body.style.zoom='100%'")
```

```python
data = driver.page_source
```

```python
soup = bs4.BeautifulSoup(data)
```

```python
soup
```

```python
all_product = soup.find_all('div',{'class':'_10Wbs- _2STCsK _3IqNCf'})
```

```python
list_all_product = []
for product in all_product:
    list_all_product.append(product.text)
list_all_product    
```

```python
all_price = soup.find_all('div',{'class':'zp9xm9 kNGSLn l-u0xK'})
```

```python
list_all_price = []
for price in all_price:
    list_all_price.append(price.text)
list_all_price    
```

```python
all_sell = soup.find_all('div',{'class':'_1uq9fs'})
```

```python
list_all_sell = []
for sell in all_sell:
    list_all_sell.append(sell.text)
list_all_sell   
```

```python
shopee_data = pd.DataFrame([list_all_product, list_all_price,list_all_sell])
shopee_data
```

```python
shopee_data =shopee_data.transpose()
shopee_data
```

```python
shopee_data.columns = ['Title','Price', 'Amount']
```

```python
shopee_data
```

```python
shopee_data.to_excel(r'shopeeโต๊ะปรับระดับ.xlsx')
```
