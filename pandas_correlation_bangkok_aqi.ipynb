## Pearson correlation

```python
df[['pm25', 'pm10', 'o3', 'no2', 'so2']].corr()
```

```python
dcorr=df[['pm25', 'pm10', 'o3', 'no2', 'so2']].corr()
dcorr
```

### style correlation matrix

```python
dcorr.style.background_gradient()
```

```python
dcorr.style.background_gradient(cmap='Greens')
```

### display correlation matrix with seaborn heatmap

```python
sns.heatmap(dcorr)
```

```python
sns.heatmap(dcorr, cmap='Reds', annot=True, fmt='.2f')
```

#### display lower triangle heatmap (using mask technique)

```python
mask=np.zeros_like(dcorr)
mask
```

```python
mask[np.triu_indices_from(mask)]=True
mask
```

```python
sns.heatmap(dcorr, cmap='Reds', annot=True, fmt='.2f', mask=mask, linewidth=.5)
```
